```python
df = pd.read_csv('C:/Users/Marcelo/OneDrive/Desktop/Triple Ten/Sprint-7/project_sql_result_07.csv')

print(df.head(), "\n")
print(df.info(), "\n")
```

```python
df['start_ts'] = pd.to_datetime(df['start_ts'], errors='coerce')
df['duration_seconds'] = pd.to_numeric(df['duration_seconds'], errors='coerce')
```

```python
# formar as amostras do teste (sábado "BAD" x sábado "GOOD")

sat = df[df['start_ts'].dt.dayofweek == 5].copy()

bad_sat  = sat[sat['weather_conditions'] == 'Bad']['duration_seconds'].dropna()
good_sat = sat[sat['weather_conditions'] == 'Good']['duration_seconds'].dropna()

print("Tamanhos (sábado) — Bad:", bad_sat.size, "| Good:", good_sat.size)
print("Médias (s) — Bad:", bad_sat.mean(), "| Good:", good_sat.mean())
print("Desvios-padrão (s) — Bad:", bad_sat.std(), "| Good:", good_sat.std())
```

```python
# teste de hipótese (Welche´s T-test, bicaudal)

from scipy import stats

alpha = 0.05  # nível de significância
res = stats.ttest_ind(bad_sat, good_sat, equal_var=False)  # Welch

t_stat = res.statistic
p_val  = res.pvalue

print("Welch t-test (Bad vs Good em sábados)")
print("t =", t_stat, "| p =", p_val, "| alpha =", alpha)

if p_val < alpha:
    print("Conclusão: rejeitamos H0 — há evidência de que a duração média muda em sábados chuvosos.")
else:
    print("Conclusão: não rejeitamos H0 — não há evidência suficiente de mudança na duração média em sábados chuvosos.")
```

### Teste de hipótese — duração média (Loop - O’Hare) em sábados chuvosos

**Hipóteses**
- H0: A duração média das corridas em sábados Bad é igual à de sábados Good.
- H1: A duração média difere entre sábados Bad e Good.

**Critério**
- Welch’s t-test (bicaudal), pois os grupos podem ter variâncias e tamanhos distintos.

**Nível de significância**
- ALPHA = 0,05

## Resultado do teste

- Amostra (sábados):
  - **Bad** (chuva/tempestade): *n* = 180, média = *2427.21 s*, desvio-padrão = 721.31 s
  - **Good** (demais condições): *n* = 888, média = *1999.68 s*, desvio-padrão = 759.20 s
- Diferença de médias (Bad − Good): *+427.53 s* (+- *7,13 min*)
- **Welch’s t-test (bicaudal)**:  
  - *t* = *7.1860*  
  - *p* = *6.74e-12*  

### Conclusão

Como **p < alpha, rejeitamos H0**.  
Há evidência estatística forte de que a duração média das corridas Loop -- O’Hare em sábados chuvosos difere da de sábados com tempo bom.  
Na prática, as corridas em sábados chuvosos são **mais longas em média 7 minutos** adicionais no período analisado.
